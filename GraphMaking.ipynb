```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts

np.random.seed(0)

x = np.random.rand(100)
y = 2*x + np.random.rand(100)


# sns.regplot(x=x, y=y, marker='v', line_kws={'color': 'green'})

fig, ax = plt.subplots()
fig.set_size_inches(10,6)

sns.regplot(x=x, y=y, marker='^', scatter_kws={'s':20}, line_kws={'color':'green'})

plt.xlabel("x")
plt.ylabel("y")

# Calculate slope, intercept, and R^2
slope, intercept, r, p, sterr = sts.linregress(x,y)
r_squared = r**2

ax.text(0.1, 0.9, f'y = {intercept:.2f} - {slope:.2f}x (R^2 = {r_squared:.3f})', transform=ax.transAxes
        )
```

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts

np.random.seed(0)

x = np.random.rand(100)
y = 2*x + np.random.rand(100)

# sns.regplot(x=x, y=y, marker='v', line_kws={'color': 'green'})
sns.set_palette("crest")

fig, ax = plt.subplots()
fig.set_size_inches(10,6)

sns.regplot(x=x, y=y, marker='_', scatter_kws={'s':20}, line_kws={'color':'green'})

plt.xlabel("x")
plt.ylabel("y")

# Calculate slope, intercept, and R^2
slope, intercept, r, p, sterr = sts.linregress(x,y)
r_squared = r**2

ax.text(0.1, 0.9, f'y = {intercept:.2f} - {slope:.2f}x (R² = {r_squared:.3f})', transform=ax.transAxes
        )
```
